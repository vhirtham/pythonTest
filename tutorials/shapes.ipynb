```python
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import
import matplotlib.pyplot as plt
import numpy as np
import copy

import mypackage.geometry as geo
import mypackage.point_cloud_generator as pcg
import mypackage.transformations as tr
```

```python
segment0 = geo.LineSegment.construct_from_points([0, 0], [1, 0])
segment0_data = segment0.rasterize(0.1)
plt.plot(segment0_data[0], segment0_data[1], "o")
```

```python
segment1 = geo.ArcSegment.construct_from_points([1,0], [2,-1], [2, 0])
segment1_data = segment1.rasterize(0.1)
plt.plot(segment1_data[0], segment1_data[1], "o")
```

```python
segment2 = geo.LineSegment.construct_from_points([2, -1], [2, -2])
segment3 = geo.LineSegment.construct_from_points([2, -2], [0, -2])

shape0 = geo.Shape2D([segment0, segment1, segment2, segment3])
shape0_data = shape0.rasterize(0.1)

plt.plot(shape0_data[0], shape0_data[1],"x-")
```

```python
plt.plot(shape0_data[0], shape0_data[1],"x-")
shape0.translate([-2.5, 0.5])
shape0_data = shape0.rasterize(0.1)
plt.plot(shape0_data[0], shape0_data[1],"rx-")
```

```python
shape_r =copy.deepcopy(shape0)
shape_r_data = shape_r.rasterize(0.1)
plt.plot(shape_r_data[0], shape_r_data[1],"x-")
shape_r.apply_transformation(tr.rotation_matrix_z(np.pi/2)[0:2,0:2])
shape_r_data = shape_r.rasterize(0.1)
plt.plot(shape_r_data[0], shape_r_data[1],"rx-")
```

```python
shape1 = copy.deepcopy(shape0)
shape1.reflect([1, 0])
shape1_data = shape1.rasterize(0.1)
plt.plot(shape0_data[0], shape0_data[1],"x-")
plt.plot(shape1_data[0], shape1_data[1],"rx-")
```

```python
profile = pcg.Profile([shape0, shape1])
profile_data = profile.rasterize(0.1)
plt.plot(profile_data[0], profile_data[1],"x-")
```
