```python
import pandas as pd
```

Load data

```python
col_names = ["0", "severity", "2", "3", "4", "5", "6", "7", "timestamp", "9", "10", "11"]
data = pd.read_csv("TcScopeServer.log", names=col_names)
```

Drop empty columns

```python
data.dropna(axis=1)
```

Convert time strings to datetime

```python
data['timestamp'] = pd.to_datetime(data['timestamp'])
```

Define  a timespan

```python
time_start = pd.to_datetime("2019-12-17T00:00:00Z")
time_end = pd.to_datetime("2019-12-18T00:00:00Z")
```

Extract data

```python
data_time_span = data[(data['timestamp'] > time_start) & (data['timestamp'] < time_end)]
data_time_span.head()
```

Groub by severity

```python
data_group_severity = data.groupby("severity")
data_group_severity.size().sort_values(ascending=False)
```

```python
data_group_severity.count()
```

Split data by severity

```python
data_error = data_group_severity.get_group("Error")
data_warning = data_group_severity.get_group("Warning")
data_information = data_group_severity.get_group("Information")
```
